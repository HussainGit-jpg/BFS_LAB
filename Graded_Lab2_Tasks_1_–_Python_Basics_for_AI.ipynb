```
class StudentOptimized:
    __slots__ = ['student_id', 'student_name', 'student_class']

    def __init__(self, student_id, student_name, student_class):
        self.student_id = student_id
        self.student_name = student_name
        self.student_class = student_class

# Create object
s2 = StudentOptimized(101, "Ali", "BSCS")

# Check memory
import sys
print("Optimized Class Memory:", sys.getsizeof(s2))
```

# 10. Apply techniques to reduce memory usage and observe the differences

```
import sys

# Normal Class
class Student:
    def __init__(self, student_id, student_name, student_class):
        self.student_id = student_id
        self.student_name = student_name
        self.student_class = student_class

# Optimized Class using __slots__
class StudentOptimized:
    __slots__ = ['student_id', 'student_name', 'student_class']

    def __init__(self, student_id, student_name, student_class):
        self.student_id = student_id
        self.student_name = student_name
        self.student_class = student_class


# Create objects
s1 = Student(101, "Ali", "BSCS")
s2 = StudentOptimized(101, "Ali", "BSCS")


# Memory comparison
print("----- Normal Class -----")
print("Object Size:", sys.getsizeof(s1))
print("Dictionary Size:", sys.getsizeof(s1.__dict__))
print("Total Memory:", sys.getsizeof(s1) + sys.getsizeof(s1.__dict__))

print("\n----- Optimized Class (__slots__) -----")
print("Object Size:", sys.getsizeof(s2))

# Check __dict__
print("\nHas __dict__?")
print("s1:", hasattr(s1, '__dict__'))
print("s2:", hasattr(s2, '__dict__'))
```
